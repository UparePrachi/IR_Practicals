```python
import pandas as pd
import numpy as np
import string
from nltk.corpus import stopwords
```

```python
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))
```

```python
for w in stopwords.words('english'):
    print(w)
    
def remove_stopwords(text):
    return " ".join([w for w in text.split() if w not in stopwords.words('english')])
```

```python
file = open('D1.txt', encoding='utf8')
doc1 = file.read()

doc1=remove_stopwords(doc1)
doc1
```

```python
file = open('D2.txt', encoding='utf8')
doc2 = file.read()
doc2

doc2=remove_stopwords(doc2)
doc2
```

```python
file = open('D3.txt', encoding='utf8')
doc3 = file.read()

doc3=remove_stopwords(doc3)
doc3
```

```python
file = open('D4.txt', encoding='utf8')
doc4 = file.read()

doc4=remove_stopwords(doc4)
doc4
```

```python
tokens1 = remove_punctuation(doc1.lower()).split()
tokens2 = remove_punctuation(doc2.lower()).split()
tokens3 = remove_punctuation(doc3.lower()).split()
tokens4 = remove_punctuation(doc4.lower()).split()

terms = list(set(tokens1 + tokens2+tokens3+tokens4))
terms
```

```python
inv_index = {}

for term in terms:
    documents = []
    if term in tokens1:
        documents.append("Document 1")
    if term in tokens2:
        documents.append("Document 2")
    if term in tokens3:
        documents.append("Document 3")
    if term in tokens4:
        documents.append("Document 4")    
        inv_index[term] = documents
```

```python
for term, documents in inv_index.items():
    print(term,"\t", "->",", ".join(documents))
```
